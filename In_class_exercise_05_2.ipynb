```python
import pandas as pd
import matplotlib.pyplot as plt
train_data=pd.read_fwf('stsa-train.txt', header=None)
test_data=pd.read_fwf('stsa-test.csv.txt',header=None)
x_train=train_data.iloc[:,1]
y_train=train_data.iloc[:,0]
x_test=test_data.iloc[:,1]
y_test=test_data.iloc[:,0]

from sklearn.model_selection import cross_val_score,KFold,cross_val_predict
kf = KFold(n_splits=10)
```

```python
x_train
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
X_train_counts = count_vect.fit_transform(x_train)
print(X_train_counts.shape)
model=MultinomialNB()
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)

print(X_train_tfidf.shape)
score=cross_val_score(model,X_train_tfidf,y_train,cv=kf)
model.fit(X_train_tfidf,y_train)
```

```python
print("Cross Validation Scores are {}".format(score))
print("Average Cross Validation score :{}".format(score.mean()))
```

```python
X_new_counts = count_vect.transform(x_test)
X_new_tfidf = tfidf_transformer.transform(X_new_counts)

predicted = model.predict(X_new_tfidf)
```

```python
import numpy as np
np.mean(predicted == y_test)
```

```python
from sklearn.linear_model import SGDClassifier
model1=SGDClassifier()
score1=cross_val_score(model1,X_train_tfidf,y_train,cv=kf)
model1.fit(X_train_tfidf,y_train)
```

```python
print("Cross Validation Scores are {}".format(score1))
print("Average Cross Validation score :{}".format(score1.mean()))
```

```python
predicted2 = model1.predict(X_new_tfidf)
np.mean(predicted2 == y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier 
model2= KNeighborsClassifier(n_neighbors=2)
score2=cross_val_score(model2,X_train_tfidf,y_train,cv=kf)
model2.fit(X_train_tfidf,y_train)
```

```python
print("Cross Validation Scores are {}".format(score2))
print("Average Cross Validation score :{}".format(score2.mean()))
predicted3 = model2.predict(X_new_tfidf)
np.mean(predicted3 == y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
model3 = DecisionTreeClassifier(random_state=0)
score3=cross_val_score(model3,X_train_tfidf,y_train,cv=10)
model3.fit(X_train_tfidf,y_train)
print("Cross Validation Scores are {}".format(score3))
print("Average Cross Validation score :{}".format(score3.mean()))
predicted4 = model3.predict(X_new_tfidf)
np.mean(predicted4 == y_test)

```

```python
#Write your code here.
import pandas as pd
data=pd.read_csv('data.csv')
pdata=pd.DataFrame()

pdata['Review']=data['Reviews'].apply(lambda x : x)
pdata['Rating']=data['Rating'].apply(lambda x : x)
```

```python
pdata.head()
```

```python
def partition(x):
  if x < 3:
    return 'Negative'
  if x==3:
    return 'Neutral'
  return 'Positive'
actualScore = pdata['Rating']
positiveNegative = actualScore.map(partition) 
pdata['Rating'] = positiveNegative
```

```python
pdata.head(20)
```

```python
import nltk
import re
nltk.download('stopwords')
from nltk.corpus import stopwords
sno = nltk.stem.SnowballStemmer('english')
stop=set(stopwords.words('english'))
def cleanpunc(sentence):
    cleaned = re.sub(r'[?|!|\'|"|#]',r'',sentence)
    cleaned = re.sub(r'[.|,|)|(|\|/]',r' ',cleaned)
    return  cleaned
print(stop)
```

```python
i=0
str1=' '
final_string=[]
all_positive_words=[] 
all_neutral_words=[]
all_negative_words=[] 
s=''
for sent in pdata['Review'].values:
    filtered_sentence=[]
    
    for w in sent.split():
        for cleaned_words in cleanpunc(w).split():
            if((cleaned_words.isalpha()) & (len(cleaned_words)>2)):    
                if(cleaned_words.lower() not in stop):
                    s=(sno.stem(cleaned_words.lower())).encode('utf8')
                    filtered_sentence.append(s)
                    if (pdata['Rating'].values)[i] == 'Positive': 
                        all_positive_words.append(s) #list of all words used to describe positive reviews
                    if(pdata['Rating'].values)[i] == 'Negative':
                        all_negative_words.append(s) 
                    if(pdata['Rating'].values)[i] == 'Neutral':
                        all_neutral_words.append(s)
                    continue
            else:
                continue 
    
    str1 = b" ".join(filtered_sentence) 
    final_string.append(str1)
    i+=1
```

```python
pdata['CleanedText']=final_string
pdata['CleanedText']=pdata['CleanedText'].str.decode("utf-8")
```

```python
pdata.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
bow = count_vect.fit_transform(pdata['CleanedText'].values)
bow.shape
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters = 3,init='k-means++',random_state=99)
model.fit(bow)
```

```python
labels = model.labels_
cluster_center=model.cluster_centers_
```

```python
final = pdata
final['Bow Clus Label'] = model.labels_ 
final.head(2)
```

```python
final.groupby(['Bow Clus Label'])['Review'].count()
```

```python
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = count_vect.get_feature_names()
for i in range(3):
    print("Cluster %d:" % i, end='')
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind], end='')
        print()
```

```python
for i in range(3):
    print("A review of assigned to cluster ", i)
    print("-" * 70)
    print(final.iloc[final.groupby(['Bow Clus Label']).groups[i][0]]['Review'])
    print('\n')
    print("_" * 70)
```

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=0.1, min_samples=20, metric='euclidean', metric_params=None, algorithm='auto', leaf_size=30)
dbmodel=db.fit(bow)
```

```python
labels = dbmodel.labels_
import numpy as np
final['Bow DB Clus Label'] = dbmodel.labels_ 
final.head(50)
core_samples_mask = np.zeros_like(db.labels_, dtype=bool)
core_samples_mask[db.core_sample_indices_] = True
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
  
print(n_clusters_)
```
